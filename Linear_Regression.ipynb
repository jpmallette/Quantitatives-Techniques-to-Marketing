```python
# configuration 
import pandas as pd
import matplotlib.pyplot as plt
from __future__ import division # to make sure it return a float
%matplotlib inline
import statsmodels.formula.api as smf
import statsmodels.stats.outliers_influence as outlier

# read and look at the data
data = pd.read_csv('C:\Users\Terry\Desktop\Project\TQM\TP2\linearRegression.csv')

# for VIF computation
data_filter = data.drop(["fashion_innovativeness"], axis=1)  
data_mat=data_filter.as_matrix()


```

```python
# look at the data 
data.head()
```

```python
# are the types corrects
data.dtypes
```

```python
# look at the relationship between input and output.
# with this graph we can see that return_policy and buying_impulsiveness seem to have a 
# linear relationship with the response. Not fashion_innovativeness however.

fig, axs = plt.subplots(1, 3, sharey=True)
data.plot(kind='scatter', x='fashion_innovativeness', y='apparel_return', ax=axs[0], figsize=(16, 8))
data.plot(kind='scatter', x='buying_impulsiveness', y='apparel_return', ax=axs[1])
data.plot(kind='scatter', x='return_policy', y='apparel_return', ax=axs[2])
```

```python
# add categorical data --- simply to show how to write a categorical variable
data['return_policy_cat'] = "whatever"
condition1 = data['return_policy'] > 0
condition2 = data['return_policy'] == 0

data.loc[condition1, 'return_policy_cat'] = "positive"
data.loc[~condition1, 'return_policy_cat'] = "negative"
data.loc[condition2, 'return_policy_cat'] = "equal_0"

data.groupby('return_policy_cat').size()
```

```python
# create three dummy variables using get_dummies, then exclude the first dummy column
sign_dummies = pd.get_dummies(data.return_policy_cat, prefix='return_policy').iloc[:, 1:]

# concatenate the dummy variable columns onto the original DataFrame (axis=0 means rows, axis=1 means columns)
data = pd.concat([data,sign_dummies], axis=1)
data.head()
```

```python
##### validate the linear regression hypothesis/Assumption and adress potential problem.


### Removing the Additive Assumption

   ## Need to look at the Correlation between input. the correlation is moderate --- 0.347 
    # between the return_policy and buying_impulsiveness. an interaction will be created with those two input
    
data.corr()
# interaction variables
data['int_return_buying'] = data.return_policy * data.buying_impulsiveness
```

```python
### Removing the Non-linearity assumption of the input.  Base on the figure above fashion_innovativeness does not seem
# to have any kind of relationship with the Y. let's see if removing the variables improve the performance of the regression. 
# CONCLUSION : slightly better adjusted R-squared without this parameter.

# create a fitted model with all three features. 
lm = smf.ols(formula='apparel_return ~ fashion_innovativeness + buying_impulsiveness + return_policy', data=data).fit()
lm.summary()

# removing fashion 
lm = smf.ols(formula='apparel_return ~  buying_impulsiveness + return_policy', data=data).fit()
lm.summary()
```

```python
### Potential Problem

# Correlation of error terms. Nop follow a normal distribution

plt.hist(lm.norm_resid())
plt.ylabel('Count')
plt.xlabel('Normalized residuals')

# Non-constant variance of error terms. Base on the graph it seems that the variance is constant

data['normalized_residual']=lm.norm_resid()
fig, axs = plt.subplots(1, 2, sharey=True)
data.plot(kind='scatter', x='buying_impulsiveness', y='residual', ax=axs[0])
data.plot(kind='scatter', x='return_policy', y='residual', ax=axs[1])



# Outliers -- Base on the figure above it's not a problem. There are many ways to tackle outliers. removing obs, 
# replacing values with median, distribution, or using a prediction method more robust to outliers.

# High-leverage points.-- Base on the figure above it's not a problem. In situation where it's harder to detect, the 
# leverage statistics is computed.

 
```

```python
# Presence of Collinearity. To identify this problem, we need to calculate the VIF, As a rule of 
# thumb, a VIF value that exceeds 5 or 10 indicates a problematic amount of collinearity
# 
# exog : ndarray, (nobs, k_vars) design matrix with all explanatory variables, as for example used in regression
# exog_idx : int index of the exogenous variable in the columns of ex

# CONCLUSION : NO PROBLEM OF COLINEARITY

outlier.variance_inflation_factor(data_mat,1)
```

```python
# Model selection 

# with interactive term

lm = smf.ols(formula='apparel_return ~  buying_impulsiveness + return_policy + int_return_buying', data=data).fit()
 
# print a summary of the fitted model
lm.summary()

  
```

```python
# CONCLUSION : the model apparel_return ~  buying_impulsiveness + return_policy seem to be the most appropriate
```
